## Outfield Players Dataset

```python
outfield_df = df[~df['player_positions'].str.contains('GK', na=False)].copy()
```

```python
gk_cols = [col for col in outfield_df.columns if col.startswith('goalkeeping')]
gk_cols
```

```python
outfield_df.drop(columns=gk_cols, inplace=True)
```

```python
outfield_df.shape
```

```python
# checking missing values
missing_outfield = outfield_df.isnull().sum().sort_values(ascending=False)
missing_outfield[missing_outfield > 0].head(10)
```

```python
# dropping because heavy text and low value
text_cols = ['player_tags', 'player_traits']
outfield_df.drop(columns=text_cols, inplace=True)
```

#### Missing By year Function

```python
def missing_by_year(df, column):
    return (
        df.groupby('fifa_version')[column]
        .apply(lambda x: x.isnull().mean())
        .round(3)
    )
```

```python
missing_by_year(outfield_df, 'release_clause_eur')
```

```python
missing_by_year(outfield_df, 'value_eur')
```

```python
missing_by_year(outfield_df, 'league_level')
```
