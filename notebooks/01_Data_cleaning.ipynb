```python
position_cols = [
    'ls','st','rs','lw','lf','cf','rf','rw',
    'lam','cam','ram','lm','lcm','cm','rcm','rm',
    'lwb','ldm','cdm','rdm','rwb',
    'lb','lcb','cb','rcb','rb','gk'
]
```

```python
def convert_rating(val):
    try:
        if pd.isna(val):
            return np.nan
        val = str(val).strip()
        if '+' in val:
            parts = val.split('+')
            return sum(int(p) for p in parts)
        elif '-' in val:
            parts = val.split('-')
            return int(parts[0]) - int(parts[1])
        else:
            return int(val)
    except:
        return np.nan
```

```python
df[position_cols] = df[position_cols].applymap(convert_rating)
```

#### Missing By year Function

```python
def missing_by_year(df, column):
    return (
        df.groupby('fifa_version')[column]
        .apply(lambda x: x.isnull().mean())
        .round(3)
    )
```

```python
missing_by_year(outfield_df, 'mentality_composure')
```

```python
missing_by_year(outfield_df, 'release_clause_eur')
```
